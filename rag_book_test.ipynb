```python
from langchain.vectorstores import Chroma
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.llms import OpenAI
from langchain.chains import RetrievalQA
from langchain.document_loaders import TextLoader
from langchain.document_loaders import PyPDFLoader
from langchain.document_loaders import DirectoryLoader
from InstructorEmbedding import INSTRUCTOR
from langchain.embeddings import HuggingFaceInstructEmbeddings
```

## загрузка документов

```python
#loader = TextLoader('/content/БелаяГв_Часть2 (1).txt')
loader = DirectoryLoader('/content/drive/MyDrive/БелаяГв', glob="./*.txt", loader_cls=TextLoader)

documents = loader.load()
```

```python
len(documents)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
texts = text_splitter.split_documents(documents)
```

```python
len(texts)
```

```python
texts[3]
```

## Instructor Embeddings

```python
from langchain.embeddings import HuggingFaceInstructEmbeddings
instructor_embeddings = HuggingFaceInstructEmbeddings(model_name="intfloat/multilingual-e5-large",
                                                      model_kwargs={"device": "cuda"})
```

## Векторная бд

```python
!rm -rf /content/db
```

```python
persist_directory = 'db'
embedding = instructor_embeddings

vectordb = Chroma.from_documents(documents=texts,
                                 embedding=embedding,
                                 persist_directory=persist_directory)
```

```python
vectordb.persist()
vectordb = None
```

```python
vectordb = Chroma(persist_directory=persist_directory,
                  embedding_function=embedding)
```

## Make a retriever

```python
retriever = vectordb.as_retriever()
```

```python
docs = retriever.get_relevant_documents("В каком юнкерском отряде служил Николка и в какой роли он был")
```

```python
len(docs)
```

```python
retriever = vectordb.as_retriever(search_kwargs={"k": 3})
```

## Гигачат запрос

```python
!pip install gigachain -q
```

```python
from langchain.chat_models.gigachat import GigaChat

llm = GigaChat(
    credentials="",
    verify_ssl_certs=False
)
```

```python
qa_chain = RetrievalQA.from_chain_type(llm,
                                  chain_type="stuff",
                                  retriever=retriever,
                                  return_source_documents=True)
```

```python
import textwrap

def wrap_text_preserve_newlines(text, width=110):
    lines = text.split('\n')

    wrapped_lines = [textwrap.fill(line, width=width) for line in lines]
    wrapped_text = '\n'.join(wrapped_lines)

    return wrapped_text

```

```python
query = "В каком юнкерском отряде служил Николка(дай цитату) и в какой роли он был(дай цитату)"
llm_response = qa_chain(query)
response_giga=llm_response['result']
```
